```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
df_small = pd.read_csv("house_price_dataset_small.csv")
df_large = pd.read_csv("house_price_dataset_large.csv")
```

```python
print("Small dataset shape:", df_small.shape)
print("Large dataset shape:", df_large.shape)

df_small.head(), df_large.head()
```

```python
# Small dataset
X_small = df_small[['Size (sq ft)']].values
y_small = df_small['Price ($)'].values

#large dataset
X_large = df_large[['Size (sq ft)']].values
y_large = df_large['Price ($)'].values
```

```python
model_small = LinearRegression()
model_large = LinearRegression()

model_small.fit(X_small, y_small)
model_large.fit(X_large, y_large)

y_small_pred = model_small.predict(X_small)
y_large_pred = model_large.predict(X_large)
```

```python
r2_small = r2_score(y_small, y_small_pred)
r2_large = r2_score(y_large, y_large_pred)

print("SMALL DATASET RESULTS")
print(f"Slope: {model_small.coef_[0]:.2f}")
print(f"Intercept: {model_small.intercept_:.2f}")
print(f"R² Score: {r2_small:.4f}")

print("\nLARGE DATASET RESULTS")
print(f"Slope: {model_large.coef_[0]:.2f}")
print(f"Intercept: {model_large.intercept_:.2f}")
print(f"R² Score: {r2_large:.4f}")
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X_small, y_small, label="Data Points")
plt.plot(X_small, y_small_pred, linewidth=2, label="Regression Line")

plt.xlabel("House Size (sq ft)")
plt.ylabel("House Price ($)")
plt.title("Linear Regression – Small Dataset")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X_large, y_large, alpha=0.6, label="Data Points")
plt.plot(X_large, y_large_pred, linewidth=2, label="Regression Line")

plt.xlabel("House Size (sq ft)")
plt.ylabel("House Price ($)")
plt.title("Linear Regression – Large Dataset")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,5))
plt.bar(["Small Dataset", "Large Dataset"], [r2_small, r2_large])

plt.ylabel("R² Score")
plt.title("Accuracy Comparison of Linear Regression Models")
plt.grid(axis='y')
plt.show()
```

Conclusion (Based on Experimental Results):

Linear regression was applied to both the small and large house price datasets to model the relationship between house size and price. The small dataset produced an R² score of 0.9591, while the large dataset achieved an R² score of 0.9533. Although the small dataset shows a slightly higher R² value, this is likely due to overfitting caused by the limited number of data points, though the model with the larger dataset will perform better on the test data.The regression model trained on the larger dataset has a more reliable slope (199.22) compared to the small dataset (157.38), indicating a more realistic estimation of price variation per unit area. Overall, despite the marginally lower R² score, the model trained on the larger dataset is more stable and better suited for generalization, making it more reliable for real-world prediction.
